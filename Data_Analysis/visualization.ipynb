```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(20,5))
plt.plot(t, np.sin(t), 'r-', label='sin(t)')
plt.plot(t, np.cos(t), 'b--', label='cos(t)')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(nrows= 2, ncols= 1, figsize=(20,6))
ax[0].plot(t, np.sin(t), 'r-', label='sine wave of t')
ax[1].plot(t, np.cos(t), 'b--', label='cosine wave of t')
ax[0].legend()
ax[1].legend()
plt.show
```

## adjusting the labels, tickets,legend and x-axis,y-axis


```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), 'g', label='sine wave of t')
plt.xlabel('x values', fontsize=14,fontdict={'color':'green'})
plt.xlabel('y values', fontsize=14,fontdict={'color':'green'})
plt.title('Sine wave', fontsize=16,fontdict={'color':'green','weight':'bold'})
plt.show()
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), 'g', label='sine wave of t')
plt.xlabel('x values', fontsize=14,fontdict={'color':'green'})
plt.xlabel('y values', fontsize=14,fontdict={'color':'green'})
plt.title('Sine wave', fontsize=16,fontdict={'color':'green','weight':'bold'})
plt.text(4, .5, 'y = sin(x)', fontsize=10, fontdict={'color':'red'})
plt.text(1, .25, 'crest = sin(x)', fontsize=10, fontdict={'color':'red'})
plt.text(4, -.5, 'throug = sin(x)', fontsize=10, fontdict={'color':'red'})
#annotate
plt.annotate('local max', xy=(1.6,1), xytext=(1,1.5),arrowprops={'facecolor':'green'})
plt.annotate('local min', xy=(4.6,-.99), xytext=(7,-.75),arrowprops={'facecolor':'green'})
            


plt.show()
```

```python
# changing background color
fig = plt.figure(figsize=(20,6), facecolor='lightpink')
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), 'g', label='sine wave of t')
#change axis background color
ax = plt.gca()
ax.set_facecolor('lightblue')
```

```python
#scatter plot
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, marker='o', color='purple')
```

```python
#bubble plot
fig = plt.figure(figsize=(20,6))
x = np.random.randn(100)
y = np.random.randn(100)
size = np.random.randint(10,500,100)
plt.scatter(x,y, marker='3', s =size, c=size, alpha=.5, edgecolors='black', cmap='RdBu')
plt.colorbar()
plt.show
```

```python
#pie chart
 
gender_count = np.array([120,45])
gender_lebels = np.array(['Female','Male'])
#text propertries
props ={'color' : 'black', 'weight': 'bold', 'fontsize': 14}
plt.pie(gender_count, labels= gender_lebels, textprops=props)
plt.show()
```

```python
fig = plt.figure(figsize=(6,6), facecolor= None)
continent = ['Africa', 'Asia', 'Euprope', 'North America','South America','oceana']
population = [1123,3425,5445,8787,3243,8974]
tprops = {'color':'black', 'weight':'bold','fontsize':8}
wprops = {'width':0.6,'linewidth': 2}
plt.pie(population, labels=continent, autopct='%.1f%%',textprops=tprops,wedgeprops=wprops)
plt.show()
```

```python
# histograph
x = np.random.randn(1000)
plt.hist(x, bins= 30,color='blue', edgecolor='black', alpha=.5)
plt.show
```

```python
#bar plot
sales = np.array([100,200,300,400,500,600,200,150,900])
months = np.array(['jan','feb','mar','apr','may','june','july','aug','sep'])
plt.bar(x=months, height=sales)
plt.show
```

```python
#bar chart wwith text
plt.bar(x=months, height=sales)
for i in range(len(sales)):
    plt.text(x=1,y=10, s=sales[1], ha='center',fontdict={'color':'white'})
plt.show()
```

```python
plt.bar(x=months, height=sales, color='green', edgecolor='black', alpha=.5)
```

```python
colors = ['red','green','blue','purple','pink','white','yellow','black','brown']
plt.bar(x=months, height=sales, color=colors, edgecolor='black', alpha=.5)
```

```python
plt.bar(x=months,
        height=sales,
        color=colors,
        edgecolor='black',
        alpha=.5,
        hatch='\\')
```

```python
#stem plot
fig = plt.figure(figsize=(20,6))
x = np.linspace(0, 10, 50)
y = np.sin(x)
plt.stem(x, y, linefmt='r:', markerfmt='r*', basefmt='g')
plt.show()
```

```python
#hexbin plot
fig = plt.figure(figsize=(8,6))
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=25, cmap='Blues_r')
plt.colorbar()
plt.show()
```
