```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_excel('data/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows=20,
    skipfooter=2)
df.head()
```

```python
cols_to_drop = ['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drop, inplace=True)
```

```python
years = list(range(1980,2014))
```

```python
df['Total'] = df[years].sum(axis=1) #column-wise sum
df.head(3)
```

```python
df.head()
```

renaming


```python
df.rename(mapper={
        'OdName':'Country',
        'AreaName': 'Continent',
        'RegName': 'Region',
        'DevName': 'Status',
        }, axis=1, inplace=True)
df.head(3)
```

```python
df.set_index('Country', inplace=True)
df.head(4)
```

```python
df.loc['Japan']
```

```python
df.loc['Japan', years].plot(kind='line',
    title="Immigration Trend of Japan", figsize=(15,5),
    linewidth=2,color='green', marker='o')
plt.show()
```

```python
df.loc[['Japan','France'], years].T.plot(kind='line')
df.loc()
```

```python
continent_df = df.groupby(by='Continent')[years].sum()
continent_df
```

```python
df.groupby(by='Continent')['Total'].sum()
```

```python
df.groupby(by='Continent')['Total'].sum().plot(kind='pie',
    autopct='%1.1f%%',
    figsize=(7,7))
plt.show()
```
